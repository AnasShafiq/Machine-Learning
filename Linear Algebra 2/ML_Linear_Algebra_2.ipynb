```python
import numpy as np
import torch
v = np.array([3.,1.])
v
```

```python
def plot_vectors(vectors, colors):
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
plot_vectors([v],['red'])
plt.xlim(-1,5)
_=plt.ylim(-1,5)
```

```python
v= torch.tensor([3,1])
I = torch.tensor([[1,0],[0,-1]])
print(I)
Iv= torch.matmul(I,v)
print(Iv)
v==Iv
```

```python
plot_vectors([Iv,v],['black','lightblue'])
plt.xlim(-1,5)
_=plt.ylim(-3,3)
```

```python
A = torch.tensor([[-1,4],[2,-2]])
A
```

```python
Av = torch.matmul(A,v)
Av
```

```python
plot_vectors([Av,v],['black','lightblue'])
plt.xlim(-1,5)
_=plt.ylim(-1,5)
```
