Police and Robber Problem.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,40,1000) #(Start point in time, time required, the details in between)
#distance by the robber
d_r = 2.5 * t
#distance of the police
d_p = 3 * (t-5)
#Ploting the graph
fig, ax = plt.subplots()
plt.title('Bank Robber Problem')
plt.xlabel('Time(In Minutes)')
plt.ylabel('Distance(In Kilometers)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t,d_r,c='blue')
ax.plot(t,d_p,c='green')
```

L2 Norm or Finding Magnitude

(Formula used - |x|=all elements^2 and sum all)

```python
#long Methord
print ('The array we are using - ',x)
print ('Applying Formula - (33**2 + 22**2 + 11**2)**(1/2)')
answer = (33**2 + 22**2 + 11**2)**(1/2)
print ('L2 value is = ',answer )

#Short Methord
np.linalg.norm(x)
```

L1 Norm

(Formula Used - |x|= Sum of all)

```python
print ('The array we are using - ',x)
print ('Applying Formula - (33 + 22 + 11)')
answer = (np.abs(x[0]) + np.abs(x[1]) + np.abs(x[2]))
answer
```

Square L2 Norm

(Formula Used - |x|=all elements^2 and sum all)

```python
#longer Methord
print ('The array we are using - ',x)
print ('Applying Formula - (33^2 + 22^2 + 11^2)')
answer = (33**2 + 22**2 + 11**2)
print (answer)

#Short Methord
np.dot(x,x)
```

Max Norm

```python
print ('The array we are using - ',x)
print ('Applying Formula - (33^2 + 22^2 + 11^2)')
answer = np.max((np.abs(x[0]) , np.abs(x[1]) , np.abs(x[2])))
answer
```

Identity Matrix

```python
#Proving if any vector is multiplied by identity matric the result is the same vector
X = torch.tensor([[1,0,0],[0,1,0],[0,0,1]])
print(X)
Y = torch.tensor([22,33,12])
print(Y)
torch.matmul(X,Y)
```

Matrix Inversion

Example
(4b + 2c = 4 ,
-5b + -3c = -7)

```python
#we have derived w = X'*Y from (y=Xw)
X = torch.tensor([[4,2],[-5,-3.]])
print(X)
Y = torch.tensor([[4,-7.]]).T
print(Y)
#for inverse of X
Xinv=torch.inverse(X)
print(Xinv)
#for W
torch.matmul(Xinv,Y)
```
